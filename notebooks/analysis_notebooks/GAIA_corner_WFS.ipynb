Show the simulated GAIA DR2 catalog used as a PhoSim input at the locations of corner wavefront sensors.

Currently the simulated images are available for the 'high' galactic latitude.

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np
fig, axs = plt.subplots(2,2,figsize=(15,15))
ax = np.ravel(axs)

i = 0 
for sensor in ['R04_S20', 'R44_S00', 'R00_S22', 'R40_S02']:
    img = ax[i].imshow(images[sensor].T, origin='lower', vmax=1200)
    ax[i].set_title('sensor %s'%sensor, fontsize=18)
    i += 1 

plt.tight_layout()

plt.savefig('GAIA_DR2_med_corner_sensors.png', bbox_inches='tight')
```
