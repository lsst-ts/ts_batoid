## Comparing Deblending Results

```python
before_update = '/astro/store/epyc/users/brycek/Commissioning/aos/ts_phosim/notebooks/analysis_scripts/wfs.zer.sep_orig_deblend.5'
no_deblend = '/astro/store/epyc/users/suberlak/Commissioning/aos/ts_phosim/notebooks/analysis_scripts/results_before_centroid_update/singleAmpSep/wfs.zer.sep.10'
opd_truth = '/astro/store/epyc/users/suberlak/Commissioning/aos/ts_phosim/notebooks/analysis_scripts/results_after_centroid_update/singleAmpSep/opd.zer.sep.5'
#after_update = '/astro/store/epyc/users/suberlak/Commissioning/aos/ts_phosim/notebooks/analysis_scripts/results_after_centroid_update/singleAmpSep/wfs.zer.sep.5'
after_update = '/astro/store/epyc/users/brycek/Commissioning/aos/ts_phosim/notebooks/analysis_scripts/wfs.zer.sep.5'
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
before = np.genfromtxt(before_update)
opd = np.genfromtxt(opd_truth)
after = np.genfromtxt(after_update)
single = np.genfromtxt(no_deblend)
```
